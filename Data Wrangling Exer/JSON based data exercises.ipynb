```python
w = pd.read_json('data/world_bank_projects.json')      #1. Reading JSON File
w = w.countryshortname                                 #2. Building table with colum name 'countryshortname'
w = w.value_counts()                                   #3. Sourcing countries' with most projects frequency.
w.head(10)                                             #4. Narrowing projects to the first 10.
```
