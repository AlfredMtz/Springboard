## Short Exploratory Data Analysis in search  for bad data

Random Forest and Decision Trees are not very sensitive to outliers. However, I still wanted to do a quick  just to see if there was anything out there that appeared to be bad data.

Preparation:

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Open files to dataframes:

```python
# Open housing data into a dataframe
dtitles = pd.DataFrame.from_csv('Housing_data/test.csv', index_col=None)

# Open housing price data into a dataframe
dprices = pd.DataFrame.from_csv('Housing_data/sample_submission.csv', index_col=None)
  
# Merge/combine the properties' descriptions and its sales prices into a new dataframe.
h_data = pd.merge(dtitles, dprices, on='Id')
```

```python
#sns.set_style("whitegrid")
sns.boxplot(y=dprices["SalePrice"], width=0.20)
```

```python
plt.scatter(h_data['Id'],h_data['SalePrice'])
plt.xlabel("Homes' Id record number(Not a feature)")
plt.ylabel('Prices')
plt.show()
```

```python
plt.scatter(h_data['GrLivArea'],h_data['SalePrice'])
plt.xlabel('Ground living area square feet')
plt.ylabel("Homes' Prices")
plt.show()
```

```python
plt.hist(h_data['BedroomAbvGr'], bins=30)
plt.xlabel('Bedrooms')
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(h_data['SalePrice'], bins=100)
plt.xlabel("Homes' Prices Distribution")
plt.ylabel("Frequency")
plt.show()
```

 Our exploration in some key variables showed that there was nothing outstanding, so we moved on.
